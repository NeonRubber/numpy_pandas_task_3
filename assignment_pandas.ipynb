```python
# Импорт модуля Pandas с псевдонимом pd
import pandas as pd
```

```python
# Загрузка датасета
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
# Вывод уникальных значений колонки 'sex' и подсчёт количества их появлений при помощи.value_counts()
print(data['sex'].value_counts())
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
# Фильтрация строк по 'sex' при помощи маски data['sex'] == 'Male'
# Подсчёт среднего значения колонки 'age' в оставшихся строках при помощи .mean
# Округление до целого значения
print(f"Средний возраст мужчин - {round(data[data['sex'] == 'Male']['age'].mean())} лет")
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
# Фильтрация при помощи маски data['native-country'] == 'United-States'
# Подсчёт доли True (США) к False (не США) при помощи .mean
# Приведение получившегося числа к "красивому" (умножение на 100 и округление до двух знаков после точки)
print(f"Доля граждан США - {round(((data['native-country'] == 'United-States').mean() * 100), 2)}%")
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
# Группировка строк по 'salary' при помощи groupby.
# Выбор столбца 'age' из всего датасета
# Подсчёт среднего значения .mean и стандартного отклонения .s результатов в одной таблице при помощи .agg
print(data.groupby('salary')['age'].agg(['mean', 'std']))
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
# Фильтрация строк по 'salary' выше 50К
# Выбор интересующего нас столбца 'education'
# Вывод уникальных значений 'education' и подсчёт количества каждой подобной записи
print(f"Статистика образования среди людей с ЗП выше 50k: {(data[data['salary'] == '>50K'])['education'].value_counts()}")
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
# Группировка по расе и полу
# Вывод подробной статистики по 'age' при помощи .describe и загрузка её в переменную described_age_stats
print("Статистика возраста для каждой расы и каждого пола:")
described_age_stats = data.groupby(['race', 'sex'])['age'].describe()
print(described_age_stats)

# Вычленение значения 'max' по 'Male' и 'Asian-Pac-Islander' при помощи .loc
print(f"Максимальный возраст мужчин расы Asian-Pac-Islander - {described_age_stats.loc[('Asian-Pac-Islander', 'Male'), 'max']}")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Создание отдельного датасета из копии "мужской части" во избежание возникновения ошибок
men_only_copy = data[data['sex'] == 'Male'].copy()

# Фильтр начинающихся с 'Married' ячеек столбца 'martial_status'
# Работа с каждой ячейкой как со строкой осуществляется через всроенный в Pandas инструмент .str
men_only_copy['married_status'] = men_only_copy['marital-status'].str.startswith('Married')

# Создание столбца с "меткой богатства" (1 или 0) для более удобной фильтрации
men_only_copy['is_rich'] = (men_only_copy['salary'] == '>50K')

# Группировка по статусу брака и подсчёт доли богатых через .mean
print("Статистика по богатым среди женатых и неженатых:")
print(f'{men_only_copy.groupby('married_status')['is_rich'].mean()}')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Вывод процента богатых среди этой группы
max_working_hours = data['hours-per-week'].max()
print(f"Максимальное время работы в неделю - {max_working_hours} часов")

# Выделение группы "работяг" и подсчёт их количества
hardest_workers = data[data['hours-per-week'] == max_working_hours]
print(f"Количество трудоголиков - {len(hardest_workers)}")

# Вычленение максимального значения из 'hours-per-week'
print(f"Процент богатых среди трудоголиков - {(hardest_workers['salary'] == '>50K').mean() * 100:.1f}%")
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
# Группировка сначала по странам, а потом - по уровню зарплат
# Отбрасывание всех колонок кроме 'hours-per-week'
# Вычисление среднего количества часов по каждой зарплатной группе
# Переформатирование таблицы при помощи .unstack в более удобный вид
print(data.groupby(['native-country', 'salary'])['hours-per-week'].mean().unstack())
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
# Создание новой колонки 'AgeGroups'
# Внесение в неё результатов подсчётов функции .cut, в качестве аргументов которой - требуемые возрастные категории
data['AgeGroup'] = pd.cut(data['age'], bins=[0, 35, 70, 100], labels=['young', 'adult', 'retiree'])

# Вывод первых десяти записей датасета с возрастом и присвоенной возрастной категорией
print(data[['age', 'AgeGroup']].head(10))
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
# Фильтрация ранее созданной колонки 'AgeGroup' по значению 'salary' == '>50K'
# Подсчёт количества людей с ЗП выше 50К
print("Количество богатых по возрастным группам:")
print((data[data['salary'] == '>50K'])['AgeGroup'].value_counts())

# Составление сводной таблицы wealth_statistics из строк 'AgeGroup' и 'salary' при помощи .crosstab
# Вывод категории, к которой относится максимальное значение 'salary'
wealth_statistics = pd.crosstab(data['AgeGroup'], data['salary'], normalize='index')
print(f"Возрастная группа, в которой чаще всего зарабатывают больше 50K - {wealth_statistics['>50K'].idxmax()}")
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
# Вывод всех профессий с суммарным количеством занятых в соответствующих профессиях
print('Все профессии с общим количеством занятых:')
print(data['occupation'].value_counts())

# Объявление функции-фильтра для фильтрафции профессий по требуемым условиям
def filter_func(group):
    age_condition = group['age'].mean() <= 40
    work_time_condition = (group['hours-per-week'] > 5).all()
    return age_condition and work_time_condition

# Применение созданного фильтра к датасету для поиска искомых профессий
print("Профессии, соответствующие заданным условиям:")
print((data.groupby('occupation').filter(filter_func))['occupation'].unique())
```
