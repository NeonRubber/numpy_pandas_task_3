* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# Версия решения задачи с использованием Python
x_python_1 = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]
prod = 1
for i in range(min(len(x_python_1), len(x_python_1[0]))):
    if x_python_1[i][i] !=0:
        prod *= x_python_1[i][i]
print(f'[Python] Ответ - {prod}')

# Версия решения задачи с использованием NumPy
x_numpy_1 = np.array([[6, 0, 9], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
print(f"[NumPy] Ответ - {np.prod(np.diag(x_numpy_1)[np.diag(x_numpy_1) != 0])}")
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# Версия решения с использованием Python
x_python_4 = [2, 2, 2, 3, 3, 3, 5]
p_numbers = [x_python_4[0]]
p_instances = [1]
for i in range(1, len(x_python_4)):
    if x_python_4[i] == x_python_4[i-1]:
        p_instances[-1] += 1
    elif x_python_4[i] != x_python_4[i-1]:
        p_numbers.append(x_python_4[i])
        p_instances.append(1)
print(f"[Python] Числа, содержащиеся в векторе - {p_numbers}, а число их повторений соответственно - {instances}")

# Версия решения с использованием NumPy
x_numpy_4 = np.array([2, 2, 2, 3, 3, 3, 5])
n_numbers = x_numpy_4[(np.concatenate(([True], x_numpy_4[:-1] != x_numpy_4[1:])))]
n_instances = np.diff(np.concatenate([[-1], np.where(x_numpy_4[:-1] != x_numpy_4[1:])[0], [len(x_numpy_4) - 1]]))
print(f"[NumPy] Числа, содержащиеся в векторе - {n_numbers} а число их повторений соответственно - {n_instances}")
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
#np.random.seed(42)

X = np.random.randint(0, 10, size=(3, 2))
Y = np.random.randint(0, 10, size=(2, 2))
print(f"Выборка X содержит следующие значения:{X}")
print(f"Выборка Y содержит следующие значения:{Y}")

# Версия решения с использованием Python
distsances = []
for i_x in range(len(X)):
    for i_y in range(len(Y)):
        difference_0 = X[i_x][0] - Y[i_y][0]
        difference_1 = X[i_x][1] - Y[i_y][1]
        distsances.append(math.sqrt(difference_0**2 + difference_1**2))
print(f"[Python] Список расстояний: {distsances}")

# Версия решения с использованием NumPy
print(f"[NumPy] Список расстояний: {np.sqrt(np.sum((X[:, np.newaxis, :] - Y[np.newaxis, :, :])**2, axis=-1))}")

# Версия решения с использованием SciPy
print(f"[SciPy] Список расстояний:{cdist(X, Y)}")
```
